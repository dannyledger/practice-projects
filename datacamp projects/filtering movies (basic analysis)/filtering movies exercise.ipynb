_You work for a production company that specializes in nostalgic styles. You want to do some research on movies released in the 1990's. You'll delve into Netflix data and perform exploratory data analysis to better understand this awesome movie decade!_

#### **Task** 
Perform exploratory data analysis on the netflix_data.csv data to understand more about movies from the 1990s decade.

- What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

- A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

#### **Data**
The dataset `netflix_data.csv`, has the following column names and descriptions. 

| Column | Description | Column | Description |
|--------|-------------|--|--|
| `show_id` | The ID of the show | `date_added` | Date added to Netflix |
| `type` | Type of show | `release_year` | Year of Netflix release |
| `title` | Title of the show | `duration` | Duration of the show in minutes |
| `director` | Director of the show | `description` | Description of the show |
| `cast` | Cast of the show | `genre` | Show genre |
| `country` | Country of origin | | |


```python
# importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# read in the netflix csv as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

# show the dataframe to get my bearings 
netflix_df
```

```python
# filter the DataFrame for movies released between 1990 and 1999
movies_90s = netflix_df[(netflix_df['type'] == 'Movie') & 
                        (netflix_df['release_year'] >= 1990) & 
                        (netflix_df['release_year'] <= 1999)]

# show revised dataframe
movies_90s
```

```python
# prepped lists
short = []
standard = []

# loop through rows using .iterrows()
for index, row in movies_90s.iterrows():
    if row['duration'] < 90 and row['genre'] == 'Action':
        short.append(1)
    else:
        standard.append(1)

# total the count of each  
short_movie_count = sum(short)
standard_movie_count = sum(standard)

# use f-strings to print a clear summary
print("Filtering complete...")
print(f"Short movies: {short_movie_count}")
print(f"Standard movies: {standard_movie_count}")
```

```python
# the exercise asks for the 'most frequent' movie duration displayed as an integer 
duration = int(movies_90s['duration'].mode().iloc[0])

# the above output made clear to the user
print(f"The most freqeunt movie duration in the 90s was {duration} minutes.")
```
